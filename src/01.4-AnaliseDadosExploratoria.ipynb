```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-deep')
import seaborn as sns
```

```python
dataset = pd.read_excel('Planilha COVID-19 completa.xlsx', header=1)
```

```python
from tabulate import tabulate

# Obter o total de registros no conjunto de dados
total_registros = len(dataset)

# Criar uma lista para armazenar os percentuais de NaN e 99999 em cada coluna
percentuais = []

# Iterar sobre todas as colunas
for coluna in dataset.columns:
    # Calcular o número e o percentual de registros NaN
    num_nan = dataset[coluna].isna().sum()
    percentual_nan = (num_nan / total_registros) * 100

    # Calcular o número e o percentual de registros com valor 99999
    num_99999 = (dataset[coluna] == 99999).sum()
    percentual_99999 = (num_99999 / total_registros) * 100

    # Adicionar os percentuais à lista
    percentuais.append([coluna, percentual_nan, percentual_99999])

# Configurar o cabeçalho da tabela
header = ['Coluna', 'NaN', '99999']

# Exibir os percentuais de NaN e 99999 em formato de tabela
print(tabulate(percentuais, headers=header, floatfmt=".2f", tablefmt='grid'))
```

```python
# Criar uma cópia do dataset
dataset_copy = dataset.copy()

# Atualizar as colunas com NaN para o conteúdo 99999
dataset_copy.fillna(99999, inplace=True)

# Obter o total de registros no conjunto de dados
total_registros = len(dataset_copy)

# Criar uma lista para armazenar os percentuais de NaN e 99999 em cada coluna
percentuais = []

# Lista para armazenar colunas atualizadas
colunas_atualizadas = []

# Iterar sobre todas as colunas
for coluna in dataset_copy.columns:
    # Calcular o número e o percentual de registros NaN
    num_nan = dataset_copy[coluna].isna().sum()
    percentual_nan = (num_nan / total_registros) * 100

    # Calcular o número e o percentual de registros com valor 99999
    num_99999 = (dataset_copy[coluna] == 99999).sum()
    percentual_99999 = (num_99999 / total_registros) * 100

    # Adicionar os percentuais à lista
    percentuais.append([coluna, percentual_nan, percentual_99999])

    # Verificar se a coluna original contém algum valor NaN e adicioná-la à lista de colunas atualizadas
    if dataset[coluna].isna().any():
        colunas_atualizadas.append(coluna)

# Configurar o cabeçalho da tabela
header = ['Coluna', 'NaN', '99999']

# Exibir os percentuais de NaN e 99999 em formato de tabela
print(tabulate(percentuais, headers=header, floatfmt=".2f", tablefmt='grid'))

# Exibir as colunas que foram atualizadas de NaN para 99999
print("\nColunas atualizadas de NaN para 99999:")
for coluna in colunas_atualizadas:
    print(coluna)
```

```python
# Agrupa os dados por 'Sexo' e 'Classif. OMS' e calcula a frequência absoluta
agrupado = dataset.groupby(['Sexo', 'Classif. OMS']).size().reset_index(name='Frequência Absoluta')

# Calcula o percentual de registros para cada combinação de 'Sexo' e 'Classif. OMS'
total = agrupado['Frequência Absoluta'].sum()
agrupado['Percentual'] = (agrupado['Frequência Absoluta'] / total) * 100

# Pivot table para obter um formato adequado para o gráfico de barras empilhadas
pivot_agrupado = agrupado.pivot_table(values='Percentual', index='Sexo', columns='Classif. OMS', fill_value=0)

# Cria o gráfico de barras empilhadas
fig, ax = plt.subplots(figsize=(10, 7))
bar_plot = pivot_agrupado.plot(kind='bar', stacked=True, ax=ax)

ax.set_title('Comparativo do percentual de registros entre Sexo e Classif. OMS')
ax.set_xlabel('Sexo')
ax.set_ylabel('Percentual de registros')
ax.legend(title='Classif. OMS', bbox_to_anchor=(1, 1), loc='upper left')

# Função para adicionar rótulos de porcentagem no topo de cada segmento da barra empilhada
def autolabel(rects, labels, ypos):
    for i, (rect, label) in enumerate(zip(rects, labels)):
        height = rect.get_height()
        if label > 0:  # Adiciona o rótulo somente se o percentual for maior que zero
            ax.text(rect.get_x() + rect.get_width()/2., ypos[i] + height,
                    '{:.2f}%'.format(label),
                    ha='center', va='bottom', fontsize=10)

# Adiciona os rótulos de porcentagem no gráfico
y_offset = [0] * len(pivot_agrupado.index)
for column, row in pivot_agrupado.iteritems():
    rects = bar_plot.containers[len(bar_plot.containers) - 1]
    autolabel(rects, row.values, y_offset)
    y_offset = [y_offset[i] + row.values[i] for i in range(len(y_offset))]

plt.show()

# Formata a tabela pivot para exibir os percentuais com duas casas decimais
formatted_table = pivot_agrupado.applymap(lambda x: '{:.2f}%'.format(x))
print(formatted_table)

from docx import Document
from docx.shared import Pt

# Cria um documento
doc = Document()

# Adiciona um título
doc.add_heading('Tabela de Percentuais', level=1)

# Adiciona uma tabela ao documento com apenas o cabeçalho
table = doc.add_table(rows=1, cols=len(formatted_table.columns) + 1)  # +1 for the index (Sexo column)

# Define a fonte do documento
style = doc.styles['Normal']
font = style.font
font.name = 'Arial'
font.size = Pt(12)

# Definindo o cabeçalho da tabela
for i, name in enumerate(["Sexo"] + list(formatted_table.columns)):
    table.cell(0, i).text = str(name)

# Preenchendo a tabela com os dados de formatted_table
for i, row in enumerate(formatted_table.iterrows()):
    cells = table.add_row().cells  # add a new row to the table and get the cells
    cells[0].text = str(row[0])  # the index (Sexo)
    for j, value in enumerate(row[1]):
        cells[j+1].text = str(value)

# Salva o documento
doc.save('Tabela de Percentuais.docx')



```

```python
# Cria grupos de idade com intervalos de 5 anos
bins = np.arange(0, dataset['Idade (anos)'].max() + 5, 5)

# Calcula a frequência e o percentual de cada grupo de idade
frequencias, _ = np.histogram(dataset['Idade (anos)'].dropna(), bins=bins)
total = len(dataset['Idade (anos)'].dropna())
percentuais = (frequencias / total) * 100

# Definir cores e categorias
colors = plt.cm.tab20(np.linspace(0, 1, len(bins) - 1))
categories = [f"{bins[i]}-{bins[i + 1]}" for i in range(len(bins) - 1)]

# Criar gráfico de barras
plt.figure(figsize=(12, 6))
bar_plot = plt.bar(categories, frequencias, color=colors)

# Adicionar rótulos de percentual no gráfico
for i, freq in enumerate(frequencias):
    height = freq
    x_position = i
    y_position = height + 0.5
    label = "{:.2f}%".format(percentuais[i])
    plt.text(x_position, y_position, label, ha='center', va='bottom', fontsize=8)

# Adicionar legenda colorida ao gráfico
legend_elements = [plt.Line2D([0], [0], color=colors[i], lw=4, label=category) for i, category in enumerate(categories)]
plt.legend(handles=legend_elements, title="Idade (anos)", bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Gráfico de barras por Idade (anos)')
plt.xticks(rotation=45)
plt.show()
```

```python
idades = dataset['Idade (anos)'].dropna()

# Calcula a média, maior e menor idade
media_idade = np.mean(idades)
maior_idade = np.max(idades)
menor_idade = np.min(idades)

# Calcula os percentuais acima e abaixo da média de idade
total = len(idades)
acima_media = np.sum(idades > media_idade)
abaixo_media = np.sum(idades <= media_idade)
percentual_acima_media = (acima_media / total) * 100
percentual_abaixo_media = (abaixo_media / total) * 100

# Configurar os dados e rótulos para o gráfico de barras
categorias = ['Acima da média', 'Abaixo da média']
percentuais = [percentual_acima_media, percentual_abaixo_media]
cores = plt.cm.tab20(np.linspace(0, 1, len(categorias)))

# Criar gráfico de barras
plt.figure(figsize=(8, 6))
bar_plot = plt.bar(categorias, percentuais, color=cores)

# Adicionar rótulos de percentual no gráfico
for i, percentual in enumerate(percentuais):
    x_position = i
    y_position = percentual + 0.5
    label = "{:.2f}%".format(percentual)
    plt.text(x_position, y_position, label, ha='center', va='bottom', fontsize=8)

# Adicionar informações sobre média, maior e menor idade
info_text = f"Média de idade: {media_idade:.2f}\nMaior idade: {maior_idade}\nMenor idade: {menor_idade}"
plt.annotate(info_text, xy=(0.5, 0.75), xycoords='axes fraction', fontsize=10, ha='center', bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white"))

plt.ylabel('Percentual')
plt.title('Percentual de pessoas acima e abaixo da média de idade')
plt.show()
```
