```python
from tabulate import tabulate

# Obter o total de registros no conjunto de dados
total_registros = len(dataset)

# Criar uma lista para armazenar os percentuais de NaN e 99999 em cada coluna
percentuais = []

# Iterar sobre todas as colunas
for coluna in dataset.columns:
    # Calcular o número e o percentual de registros NaN
    num_nan = dataset[coluna].isna().sum()
    percentual_nan = (num_nan / total_registros) * 100

    # Calcular o número e o percentual de registros com valor 99999
    num_99999 = (dataset[coluna] == 99999).sum()
    percentual_99999 = (num_99999 / total_registros) * 100

    # Adicionar os percentuais à lista
    percentuais.append([coluna, percentual_nan, percentual_99999])

# Configurar o cabeçalho da tabela
header = ['Coluna', 'NaN', '99999']

# Exibir os percentuais de NaN e 99999 em formato de tabela
print(tabulate(percentuais, headers=header, floatfmt=".2f", tablefmt='grid'))
```
