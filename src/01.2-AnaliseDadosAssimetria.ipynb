```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-deep')
import seaborn as sns
```

```python
dataset = pd.read_excel('Planilha COVID-19 completa.xlsx', header=1)
```

```python
from scipy.stats import skew, kurtosis

for col in dataset.columns:
    if dataset[col].dtype != 'object':  # se a coluna não for categórica
        assymetry = skew(dataset[col].dropna())  # calcular assimetria
        kurt = kurtosis(dataset[col].dropna())  # calcular curtose
        print(f'Coluna: {col} | Assimetria: {assymetry:.2f} | Curtose: {kurt:.2f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in dataset.columns:
    if dataset[col].dtype != 'object':  # se a coluna não for categórica
        plt.figure(figsize=(12, 6))
        sns.histplot(dataset[col].dropna(), kde=True)
        plt.title(f'Histograma da coluna: {col}')
        plt.show()
```

```python
for column in dataset.columns:
    print(f"Coluna: {column}")
    value_counts = dataset[column].value_counts()
    modes = value_counts[value_counts == value_counts.max()].index.tolist()
    print(f"Modos: {modes}\n")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Inicialize duas listas vazias para armazenar os valores de assimetria e curtose
assymetry_values = []
kurt_values = []

# Inicialize uma lista vazia para armazenar os nomes das colunas
col_names = []

for col in dataset.columns:
    if dataset[col].dtype != 'object':  # se a coluna não for categórica
        assymetry = skew(dataset[col].dropna())  # calcular assimetria
        kurt = kurtosis(dataset[col].dropna())  # calcular curtose
        assymetry_values.append(assymetry)
        kurt_values.append(kurt)
        col_names.append(col)

# Crie um gráfico de barras para a assimetria
plt.figure(figsize=(20, 10))
plt.bar(col_names, assymetry_values)
plt.xticks(rotation=90)  # Rotacione os rótulos do eixo x para evitar sobreposição
plt.title('Assimetria por coluna')
plt.show()

# Crie um gráfico de barras para a curtose
plt.figure(figsize=(20, 10))
plt.bar(col_names, kurt_values)
plt.xticks(rotation=90)  # Rotacione os rótulos do eixo x para evitar sobreposição
plt.title('Curtose por coluna')
plt.show()
```

```python
from docx import Document
from docx.shared import Pt
from scipy.stats import skew, kurtosis

# Criando um novo Documento
doc = Document()

# Adicionando um título ao documento
doc.add_heading('Assimetria e Curtose', 0)

# Adicionando uma tabela ao documento
table = doc.add_table(rows=1, cols=3)

# Definindo o cabeçalho da tabela
for i, name in enumerate(["Coluna", "Assimetria", "Curtose"]):
    table.cell(0, i).text = name

# Preenchendo a tabela com os valores de assimetria e curtose de cada coluna
for col in dataset.columns:
    if dataset[col].dtype != 'object':  # se a coluna não for categórica
        assymetry = skew(dataset[col].dropna())  # calcular assimetria
        kurt = kurtosis(dataset[col].dropna())  # calcular curtose

        # Adicionando uma nova linha na tabela para cada coluna
        cells = table.add_row().cells
        cells[0].text = col
        cells[1].text = f'{assymetry:.2f}'
        cells[2].text = f'{kurt:.2f}'

# Salvando o documento
doc.save('Assimetria_Curtose.docx')
```
