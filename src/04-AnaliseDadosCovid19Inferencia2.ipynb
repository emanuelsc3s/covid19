```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, accuracy_score

modelLR = LogisticRegression(random_state=1)

# Aqui vou guardar as matrizes de confusão
conf_matrix_list_of_arrays = []

# Usar kfold para separar os índices de treino e de teste
kf = KFold(n_splits=10, random_state=10, shuffle=True)
# Guardar as precisões de 10 rodadas
precisao = []

# loop das 5 rodadas
for train_index, test_index in kf.split(previsores):
  #Separação em treino e teste
  X_train, X_test = previsores[train_index], previsores[test_index]
  y_train, y_test = classe[train_index], classe[test_index]
  # Treinar o modelo
  modelLR.fit(X_train, y_train)
  # Calcular a matriz de confusão
  conf_matrix = confusion_matrix(y_test,modelLR.predict(X_test))
  # Guardar a matriz de confusão
  conf_matrix_list_of_arrays.append(conf_matrix)
  # Guardar a precisão
  precisao.append( accuracy_score(y_test, modelLR.predict(X_test)) )

# imprimir o array de precisões
print(precisao)
# imprimir as matrizes de confusão (nesse caso estou imprimindo a matriz que o modelo teve a melhor acurácia)
print(conf_matrix_list_of_arrays[1])
```

```python
# tive que rotular o array classe para poder treinar a rede neural 
from sklearn.preprocessing import LabelBinarizer

train_y = classe
lb = LabelBinarizer()
lb.fit(train_y)
classe = lb.transform(train_y)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, accuracy_score

# rede neural (nesse ponto tem que experimentar outras arquiteturas da rede #hidden_layer_sizes)
model = MLPRegressor(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(200,300,400,300,200), max_iter = 100000, random_state=1)

# Aqui vou guardar as matrizes de confusão
conf_matrix_list_of_arrays = []
# Usar kfold para separar os índices de treino e de teste
kf = KFold(n_splits=5, random_state=10, shuffle=True)
# Guardar as precisões de 5 rodadas
precisao = []

# loop das 5 rodadas
for train_index, test_index in kf.split(previsores):
  #Separação em treino e teste
  X_train, X_test = previsores[train_index], previsores[test_index]
  y_train, y_test = classe[train_index], classe[test_index]
  # Treinar o modelo
  model.fit(X_train, y_train)
  # Calcular a matriz de confusão
  conf_matrix = confusion_matrix(lb.inverse_transform(y_test),lb.inverse_transform(model.predict(X_test)))
  # Guardar a matriz de confusão
  conf_matrix_list_of_arrays.append(conf_matrix)
  # Guardar a precisão
  precisao.append( accuracy_score(lb.inverse_transform(y_test), lb.inverse_transform(model.predict(X_test))) )

# imprimir o array de precisões
print(precisao)
# imprimir as matrizes de confusão (nesse caso estou imprimindo a matriz que o modelo teve a melhor acurácia)
print(conf_matrix_list_of_arrays[1])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, accuracy_score

# rede neural (nesse ponto tem que experimentar outras arquiteturas da rede #hidden_layer_sizes)
modelRF = RandomForestClassifier(max_depth=3, random_state=0) 

# Aqui vou guardar as matrizes de confusão
conf_matrix_list_of_arrays = []
# Usar kfold para separar os índices de treino e de teste
kf = KFold(n_splits=5, random_state=10, shuffle=True)
# Guardar as precisões de 5 rodadas
precisao = []

# loop das 5 rodadas
for train_index, test_index in kf.split(previsores):
  
  #Separação em treino e teste - separar 70% / 30%
  X_train, X_test = previsores[train_index], previsores[test_index]
  y_train, y_test = classe[train_index], classe[test_index]
  # Treinar o modelo
  modelRF.fit(X_train, y_train)
  
  
  # Calcular a matriz de confusão - retirar
  conf_matrix = confusion_matrix(lb.inverse_transform(y_test),lb.inverse_transform(modelRF.predict(X_test)))
  # Guardar a matriz de confusão
  
  conf_matrix_list_of_arrays.append(conf_matrix)
  # Guardar a precisão
  precisao.append( accuracy_score(lb.inverse_transform(y_test), lb.inverse_transform(modelRF.predict(X_test))) )

# imprimir o array de precisões - add demais metricas
print(precisao)

# imprimir as matrizes de confusão (nesse caso estou imprimindo a matriz que o modelo teve a melhor acurácia)
print(conf_matrix_list_of_arrays[1])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve

def plot_roc_curve(fper, tper):
    plt.plot(fper, tper, color='red', label='ROC')
    plt.plot([0, 1], [0, 1], color='green', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic Curve')
    plt.legend()
    plt.show()
    
data_X, cls_lab = make_classification(n_samples=2100, n_classes=2, weights=[1,1], random_state=2)
train_X, test_X, train_y, test_y = train_test_split(data_X, cls_lab, test_size=0.5, random_state=2)

model =RandomForestClassifier()
model.fit(train_X, train_y)

prob = model.predict_proba(test_X)
prob = prob[:, 1]
fper, tper, thresholds = roc_curve(test_y, prob)
plot_roc_curve(fper, tper)
```
